```python
!pip install soundfile librosa
import soundfile as sf
import io
from tqdm import tqdm
from datasets import load_dataset
import librosa
import pandas as pd
import numpy as np

def load_existing_chunks(output_dir=None):
    """
    Load existing audio chunks from YT directory
    
    Args:
        output_dir: Directory where chunks are stored (default: {home}/audio_chunks_6s)
    
    Returns:
        DataFrame with paths and labels
    """
    
    if output_dir is None:
        output_dir = f"{home}/audio_chunks_6s"
    
    print(f"Loading existing chunks from: {output_dir}")
    
    chunk_paths = []
    chunk_labels = []
    
    # Load real chunks
    real_dir = f"{output_dir}/real"
    if yt.exists(real_dir):
        real_files = yt.list(real_dir)
        for file in real_files:
            if file.endswith('.wav'):
                chunk_paths.append(f"{real_dir}/{file}")
                chunk_labels.append(1)  # real = 1
        print(f"Found {len(real_files)} real chunks")
    
    # Load fake chunks
    fake_dir = f"{output_dir}/fake"
    if yt.exists(fake_dir):
        fake_files = yt.list(fake_dir)
        n_fake_before = len(chunk_paths)
        for file in fake_files:
            if file.endswith('.wav'):
                chunk_paths.append(f"{fake_dir}/{file}")
                chunk_labels.append(0)  # fake = 0
        print(f"Found {len(chunk_paths) - n_fake_before} fake chunks")
    
    # Create dataframe
    chunks_df = pd.DataFrame({
        'path': chunk_paths,
        'is_fake': chunk_labels
    })
    
    print(f"\nTotal chunks loaded: {len(chunks_df)}")
    print(f"  - Real: {sum(chunks_df['is_fake'] == 1)}")
    print(f"  - Fake: {sum(chunks_df['is_fake'] == 0)}")
    
    return chunks_df

def download_and_split_to_chunks(dataset_name, token, chunk_duration=6.0, target_sr=16000, output_dir=None):
    """
    Download HF dataset and directly split into chunks without saving full files
    
    Args:
        dataset_name: HuggingFace dataset name
        token: HuggingFace token
        chunk_duration: Duration of each chunk in seconds (default: 6.0)
        target_sr: Target sample rate (default: 16000)
        output_dir: Output directory in YT
    
    Returns:
        DataFrame with chunk paths and labels
    """
    
    if output_dir is None:
        output_dir = f"{home}/audio_chunks_{int(chunk_duration)}s"
    
    # Create YT directories
    real_dir = f"{output_dir}/real"
    fake_dir = f"{output_dir}/fake"
    
    for dir_path in [output_dir, real_dir, fake_dir]:
        if not yt.exists(dir_path):
            yt.create("map_node", dir_path)
    
    # Load dataset
    print("Loading dataset from HuggingFace...")
    dataset = load_dataset(dataset_name, split="train", token=token)
    
    chunk_paths = []
    chunk_labels = []
    chunk_original_idx = []
    
    chunk_samples = int(chunk_duration * target_sr)
    
    real_chunks = 0
    fake_chunks = 0
    total_chunks = 0
    
    print(f"Processing {len(dataset)} files and splitting into {chunk_duration}s chunks...")
    
    # Process each audio file
    for idx, sample in enumerate(tqdm(dataset, desc="Processing")):
        try:
            label = sample['label']  # 0=fake, 1=real
            
            # Get audio data
            audio_array = sample['audio']['array']
            sample_rate = sample['audio']['sampling_rate']
            
            # Resample if needed
            if sample_rate != target_sr:
                audio_array = librosa.resample(audio_array, orig_sr=sample_rate, target_sr=target_sr)
            
            # Calculate number of chunks
            n_chunks = max(1, int(np.ceil(len(audio_array) / chunk_samples)))
            
            # Process each chunk
            for chunk_idx in range(n_chunks):
                start = chunk_idx * chunk_samples
                end = min(start + chunk_samples, len(audio_array))
                
                # Extract chunk
                chunk = audio_array[start:end]
                
                # Pad last chunk if necessary
                if len(chunk) < chunk_samples:
                    padding = chunk_samples - len(chunk)
                    chunk = np.pad(chunk, (0, padding), mode='constant')
                
                # Convert to WAV bytes
                buffer = io.BytesIO()
                sf.write(buffer, chunk, target_sr, format='WAV')
                wav_bytes = buffer.getvalue()
                
                # Determine folder and filename
                if label == 1:
                    folder = real_dir
                    filename = f"real_{idx}_chunk_{chunk_idx}.wav"
                    real_chunks += 1
                else:
                    folder = fake_dir
                    filename = f"fake_{idx}_chunk_{chunk_idx}.wav"
                    fake_chunks += 1
                
                # Save chunk to YT
                file_path = f"{folder}/{filename}"
                yt.write_file(file_path, wav_bytes)
                
                chunk_paths.append(file_path)
                chunk_labels.append(label)
                chunk_original_idx.append(idx)
                total_chunks += 1
                
        except Exception as e:
            print(f"Error processing sample {idx}: {e}")
            continue
    
    # Create dataframe
    chunks_df = pd.DataFrame({
        'path': chunk_paths,
        'is_fake': chunk_labels
    })
    
    print(f"\n✅ Processing complete!")
    print(f"Original files: {len(dataset)}")
    print(f"Total chunks created: {total_chunks}")
    print(f"  - Real chunks: {real_chunks}")
    print(f"  - Fake chunks: {fake_chunks}")
    print(f"Average chunks per file: {total_chunks / len(dataset):.2f}")
    print(f"Files saved to: {output_dir}")
    
    return chunks_df
```

```python
!pip install pydub
import os
import pandas as pd
import torch
import torchaudio
import random
import numpy as np
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
from pathlib import Path
from tractorun.backend.tractorch import TensorSerializer
import soundfile as sf
import io
from tqdm import tqdm
from datasets import load_dataset
from pydub import AudioSegment

def balance_dataset(audio_files, labels, margin = 5000):
    # Separate by class
    real_indices = [i for i, l in enumerate(labels) if l == 1]
    fake_indices = [i for i, l in enumerate(labels) if l == 0]
    
    n_fake_target = len(real_indices) + margin
    sampled_fake = np.random.choice(fake_indices, min(n_fake_target, len(fake_indices)), replace=False)

    sampled_indices = list(real_indices) + list(sampled_fake)

    audio_files = [audio_files[i] for i in sampled_indices]
    labels = [labels[i] for i in sampled_indices]
    
    # Create DataFrame
    data_df = pd.DataFrame({
        'path': audio_files,
        'is_fake': labels
    })
    
    return data_df

def prepare_asvspoof_data(base_path):
    """Process ASVspoof 2021 dataset"""
    
    if not os.path.exists(base_path):
        raise FileNotFoundError(f"Dataset path does not exist: {base_path}")
    
    # Load metadata file
    metadata_files = [
        'DF-keys-full/keys/DF/CM/trial_metadata.txt',
        'PA-keys-full/keys/PA/CM/trial_metadata.txt',
        'LA-keys-full/keys/LA/CM/trial_metadata.txt'
    ]
    
    df_keys_paths = []
    for metadata_file in metadata_files:
        test_path = os.path.join(base_path, metadata_file)
        if os.path.exists(test_path):
            df_keys_paths.append(test_path)
    
    if df_keys_paths is None:
        raise FileNotFoundError("Metadata file not found")
    
    # Read metadata
    data_entries = []
    for df_keys_path in df_keys_paths:
        with open(df_keys_path, 'r') as f:
            for line in f:
                parts = line.strip().split()
                if len(parts) >= 6:
                    data_entries.append(parts)
    
    # Look for audio files
    flac_dirs = []
    for pattern in ['ASVspoof2021_DF_eval_part{}/ASVspoof2021_DF_eval/flac', 'ASVspoof2021_PA_eval_part{}/ASVspoof2021_PA_eval/flac', 'ASVspoof2021_LA_eval/ASVspoof2021_LA_eval/flac']:
        if '{}' in pattern:
            for part in ['00', '01', '02', '03']:
                flac_path = os.path.join(base_path, pattern.format(part))
                if os.path.exists(flac_path):
                    flac_dirs.append(flac_path)
        else:
            flac_path = os.path.join(base_path, pattern)
            if os.path.exists(flac_path):
                flac_dirs.append(flac_path)
    
    if not flac_dirs:
        raise FileNotFoundError(f"No audio directories found in {base_path}")
    
    # Create file mapping
    file_id_to_path = {}
    for flac_dir in flac_dirs:
        for file in os.listdir(flac_dir):
            if file.endswith('.flac'):
                file_id = os.path.splitext(file)[0]
                file_id_to_path[file_id] = os.path.join(flac_dir, file)
    
    # Process files and labels
    audio_files = []
    labels = []
    
    for entry in data_entries:
        file_id = entry[1]
        label = entry[5]
        
        if file_id in file_id_to_path:
            audio_files.append(file_id_to_path[file_id])
            if label.lower() in ['spoof', 'fake', 'spoofed']:
                labels.append(0)
            elif label.lower() in ['bonafide', 'real', 'genuine']:
                labels.append(1)
            else:
                continue
    
    if len(audio_files) == 0:
        raise ValueError("No audio files could be matched with metadata")

    data_df = balance_dataset(audio_files, labels)

    if yt.exists("//home/hyperplane/audio_chunks_6s"):
        print("Found existing chunks directory!")
        data_hf_df = load_existing_chunks("//home/hyperplane/audio_chunks_6s")
    else:
        data_hf_df = download_and_split_to_chunks("CosminMihai02/real_fake_audio", hf_token)

    data_df = pd.concat([data_df, data_hf_df], ignore_index=True)
    
    return data_df
```

```python
import torchaudio.transforms as transforms
import torchaudio
import random

class AudioUtil:
    @staticmethod
    def open(audio_file):
        # Modified to handle YT files
        try:
            # Try YT first
            audio_bytes = yt.read_file(audio_file).read()
            sig, sr = torchaudio.load(io.BytesIO(audio_bytes))
        except:
            # Fallback to regular file
            sig, sr = torchaudio.load(audio_file)
        return sig, sr

    @staticmethod
    def rechannel(aud, new_channel):
        sig, sr = aud
        if sig.shape[0] == new_channel:
            return aud
        if new_channel == 1:
            resig = sig[:1, :]
        else:
            resig = sig.repeat(new_channel, 1)
        return resig, sr

    @staticmethod
    def resample(aud, newsr):
        sig, sr = aud
        if sr == newsr:
            return aud
        num_channels = sig.shape[0]
        resig = transforms.Resample(sr, newsr)(sig[:1, :])
        if num_channels > 1:
            retwo = transforms.Resample(sr, newsr)(sig[1:, :])
            resig = torch.cat([resig, retwo], dim=0)
        return resig, newsr

    @staticmethod
    def pad_trunc(aud, max_ms):
        sig, sr = aud
        num_rows, sig_len = sig.shape
        max_len = sr // 1000 * max_ms
        if sig_len > max_len:
            sig = sig[:, :max_len]
        elif sig_len < max_len:
            pad_begin_len = random.randint(0, max_len - sig_len)
            pad_end_len = max_len - sig_len - pad_begin_len
            pad_begin = torch.zeros((num_rows, pad_begin_len))
            pad_end = torch.zeros((num_rows, pad_end_len))
            sig = torch.cat((pad_begin, sig, pad_end), dim=1)
        return sig, sr

    @staticmethod
    def time_shift(aud, shift_pct):
        sig, sr = aud
        _, sig_len = sig.shape
        shift_amt = int(random.random() * shift_pct * sig_len)
        return sig.roll(shift_amt), sr

    @staticmethod
    def spectro_gram(aud, n_mels=64, n_fft=1024, hop_len=None):
        sig, sr = aud
        spec = transforms.MelSpectrogram(sr, n_fft=n_fft, hop_length=hop_len, n_mels=n_mels)(sig)
        spec = transforms.AmplitudeToDB(top_db=80)(spec)
        return spec

    @staticmethod
    def spectro_augment(spec, max_mask_pct=0.1, n_freq_masks=1, n_time_masks=1):
        _, n_mels, n_steps = spec.shape
        mask_value = spec.mean()
        aug_spec = spec
        freq_mask_param = int(max_mask_pct * n_mels)
        for _ in range(n_freq_masks):
            aug_spec = transforms.FrequencyMasking(freq_mask_param)(aug_spec, mask_value)
        time_mask_param = int(max_mask_pct * n_steps)
        for _ in range(n_time_masks):
            aug_spec = transforms.TimeMasking(time_mask_param)(aug_spec, mask_value)
        return aug_spec

    @staticmethod
    def add_noise(aud, noise_level=0.005):
        sig, sr = aud
        noise = torch.randn_like(sig) * noise_level
        return sig + noise, sr

    @staticmethod
    def change_gain(aud, gain_db_range=(-6, 6)):
        sig, sr = aud
        gain_db = random.uniform(*gain_db_range)
        gain = 10 ** (gain_db / 20)
        return sig * gain, sr

    @staticmethod
    def pitch_shift(aud, n_steps=2):
        sig, sr = aud
        return torchaudio.functional.pitch_shift(sig, sr, n_steps), sr

    @staticmethod
    def get_mfcc(aud, n_mfcc=64, n_fft=1024, hop_len=None):
        """Extract MFCC using torchaudio"""
        sig, sr = aud
        mfcc_transform = transforms.MFCC(
            sample_rate=sr,
            n_mfcc=n_mfcc,
            melkwargs={
                'n_fft': n_fft,
                'hop_length': hop_len,
                'n_mels': 128  # Standard value for MFCC
            }
        )
        mfcc = mfcc_transform(sig)
        return mfcc

    @staticmethod
    def get_lfcc(aud, n_lfcc=64, n_fft=1024, hop_len=None):
        """Extract LFCC using torchaudio LFCC transform"""
        sig, sr = aud

        lfcc_transform = transforms.LFCC(
            sample_rate=sr,
            n_lfcc=n_lfcc,
            speckwargs={
                'n_fft': n_fft,
                'hop_length': hop_len
            }
        )
        lfcc = lfcc_transform(sig)
    
        return lfcc
```

```python
from torch.utils.data import DataLoader, Dataset, random_split

class MultiChannelSoundDS(Dataset):
    def __init__(self, dataframe, mode="original"):
        """
        Args:
            dataframe: pandas DataFrame with 'path' and 'is_fake' columns
            mode: augmentation mode
        """
        self.df = dataframe.reset_index(drop=True)
        self.duration = 6000  # milliseconds
        self.sr = 16000
        self.channel = 1  # mono
        self.shift_pct = 0.4
        self.mode = mode
        
        # Feature extraction parameters
        self.n_mels = 64
        self.n_mfcc = 64
        self.n_lfcc = 64
        self.n_fft = 780
        self.hop_len = 195
        
    def __len__(self):
        return len(self.df)
    
    def __getitem__(self, idx):
        row = self.df.iloc[idx]
        audio_file = row['path']
        class_id = row['is_fake']
        
        # Load and preprocess audio
        aud = AudioUtil.open(audio_file)
        aud = AudioUtil.resample(aud, self.sr)
        aud = AudioUtil.rechannel(aud, self.channel)
        aud = AudioUtil.pad_trunc(aud, self.duration)
        
        # Apply augmentations
        if self.mode == "time_shift":
            aud = AudioUtil.time_shift(aud, self.shift_pct)
        elif self.mode == "add_noise":
            aud = AudioUtil.add_noise(aud, noise_level=random.uniform(0.005, 0.05))
        elif self.mode == "pitch_shift":
            aud = AudioUtil.pitch_shift(aud, n_steps=random.randint(1, 3))
        elif self.mode == "change_gain":
            aud = AudioUtil.change_gain(aud, gain_db_range=(-6, 6))
        elif self.mode == "combined":
            aud = AudioUtil.time_shift(aud, self.shift_pct)
            aud = AudioUtil.add_noise(aud, noise_level=random.uniform(0.005, 0.05))
            if random.random() > 0.5:
                aud = AudioUtil.change_gain(aud, gain_db_range=(-6, 6))
        
        # Extract 3 channels of features
        mel_spec = AudioUtil.spectro_gram(aud, n_mels=self.n_mels, n_fft=self.n_fft, hop_len=self.hop_len)
        mfcc = AudioUtil.get_mfcc(aud, n_mfcc=self.n_mfcc, n_fft=self.n_fft, hop_len=self.hop_len)
        lfcc = AudioUtil.get_lfcc(aud, n_lfcc=self.n_lfcc, n_fft=self.n_fft, hop_len=self.hop_len)
        
        # Ensure all features have the same dimensions
        # mel_spec shape: (1, n_mels, time_frames)
        # mfcc shape: (1, n_mfcc, time_frames)
        # lfcc shape: (1, n_lfcc, time_frames)
        
        # Get the minimum time frames to ensure consistency
        min_frames = min(mel_spec.shape[-1], mfcc.shape[-1], lfcc.shape[-1])
        mel_spec = mel_spec[:, :, :min_frames]
        mfcc = mfcc[:, :, :min_frames]
        lfcc = lfcc[:, :, :min_frames]
        
        # Stack as 3-channel tensor
        features = torch.cat([mel_spec, mfcc, lfcc], dim=0)  # (3, n_features, time_frames)
        
        # Apply spectrogram augmentation if needed
        if self.mode in ["spectro_augment", "combined"]:
            features = AudioUtil.spectro_augment(
                features, 
                max_mask_pct=0.1, 
                n_freq_masks=random.randint(1, 3), 
                n_time_masks=random.randint(1, 3)
            )
        
        return features, class_id
```

```python

def batcher(iterable, batch_size=1000):
    batch = []
    for item in iterable:
        batch.append(item)
        if len(batch) == batch_size:
            yield batch
            batch = []
    if batch:
        yield batch

@yt.yt_dataclass
class AudioRow:
    data: bytes              # Use Python's str type
    labels: bytes            # Use Python's int type
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import models

class AttentionPool(nn.Module):
    def __init__(self, in_dim):
        super().__init__()
        self.attn = nn.Linear(in_dim, 1)

    def forward(self, x):
        # x: [B, T, D]
        scores = self.attn(x)              # [B, T, 1]
        weights = F.softmax(scores, dim=1) # [B, T, 1]
        return (weights * x).sum(dim=1)    # [B, D]

class CRNNWithAttn(nn.Module):
    def __init__(self,  pretrained=True, hidden_size=128, num_layers=1, dropout=0.2):
        super().__init__()
        # 1. Load pretrained ResNet18 (with 3-channel input)
        resnet = models.resnet18(weights='DEFAULT' if pretrained else None)
        # Remove final pooling & fc
        self.backbone = nn.Sequential(*list(resnet.children())[:-2])

        # 2. Bi-GRU for temporal modeling
        self.gru = nn.GRU(
            input_size=512,          # ResNet last block outputs 512 channels
            hidden_size=hidden_size,
            num_layers=num_layers,
            batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers>1 else 0.0
        )

        # 3. Attention pooling
        self.attn_pool = AttentionPool(hidden_size*2)

        # 4. Classification head
        self.classifier = nn.Sequential(
            nn.Linear(hidden_size*2, hidden_size),
            nn.ReLU(inplace=True),
            nn.Dropout(dropout),
            nn.Linear(hidden_size, 1)
        )

    def forward(self, x):
        # x: [B, 3, F, T]
        feat = self.backbone(x)            # [B, 512, F', T']
        feat = feat.mean(dim=2)            # collapse freq → [B,512,T']
        feat = feat.permute(0,2,1)         # → [B,T',512]

        out, _ = self.gru(feat)            # → [B,T',2*hidden_size]
        pooled = self.attn_pool(out)       # → [B,2*hidden_size]
        return self.classifier(pooled)     # → [B,1]
```

```python
!pip install tensorboard

from tractorun.run import run
from tractorun.backend.tractorch import Tractorch, YtTensorDataset
from tractorun.toolbox import Toolbox
from tractorun.mesh import Mesh
from tractorun.resources import Resources
from tractorun.stderr_reader import StderrMode
from tractorun.backend.tractorch.serializer import TensorSerializer

from torch.utils.tensorboard import SummaryWriter
from torch.utils.data import DataLoader
from tqdm.auto import tqdm
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import argparse


# ----------------------------
# Train Function
# ----------------------------
def train_one_epoch(model, train_dl, optimizer, scheduler, criterion, device, epoch, num_epochs):
    model.train()
    epoch_bar = tqdm(train_dl, desc=f"Epoch {epoch}/{num_epochs}", unit="batch")

    running_loss, correct_preds, total_preds = 0.0, 0, 0

    imagenet_mean = torch.tensor([0.485, 0.456, 0.406], device=device).view(1, 3, 1, 1)
    imagenet_std  = torch.tensor([0.229, 0.224, 0.225], device=device).view(1, 3, 1, 1)

    for step, (inputs, labels) in enumerate(epoch_bar):
        inputs, labels = inputs.to(device), labels.to(device).unsqueeze(1).float()
        inputs = (inputs - imagenet_mean) / imagenet_std

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        scheduler.step()

        running_loss += loss.item()
        preds = (torch.sigmoid(outputs) > 0.5).float()
        correct_preds += (preds == labels).sum().item()
        total_preds += preds.size(0)

        epoch_bar.set_postfix({
            "loss": f"{running_loss / (step+1):.4f}",
            "acc":  f"{correct_preds / total_preds:.4f}"
        })

    avg_loss = running_loss / len(train_dl)
    avg_acc = correct_preds / total_preds

    return avg_loss, avg_acc

# ----------------------------
# Validation Function
# ----------------------------
def validate(model, val_dl, criterion, device, epoch):
    model.eval()
    val_loss, val_correct, val_total = 0.0, 0, 0

    imagenet_mean = torch.tensor([0.485, 0.456, 0.406], device=device).view(1, 3, 1, 1)
    imagenet_std  = torch.tensor([0.229, 0.224, 0.225], device=device).view(1, 3, 1, 1)

    with torch.no_grad():
        for inputs, labels in val_dl:
            inputs, labels = inputs.to(device), labels.to(device).unsqueeze(1).float()
            inputs = (inputs - imagenet_mean) / imagenet_std

            outputs = model(inputs)
            val_loss += criterion(outputs, labels).item()
            preds = (torch.sigmoid(outputs) > 0.5).float()
            val_correct += (preds == labels).sum().item()
            val_total += preds.size(0)

    avg_val_loss = val_loss / len(val_dl)
    avg_val_acc = val_correct / val_total

    return avg_val_loss, avg_val_acc

```

```python
# ----------------------------
# Main entrypoint
# ----------------------------
def main(toolbox: Toolbox):
    args = argparse.Namespace(batch_size=32, num_epochs=10, patience=3, log_dir=f"{working_dir}/runs/exp1")

    # YT datasets
    dataset1 = YtTensorDataset(path=dataset_train_path, yt_client=toolbox.yt_client, columns=['data', 'labels'])
    dataset2 = YtTensorDataset(path=dataset_test_path, yt_client=toolbox.yt_client, columns=['data', 'labels'])

    train_dl = DataLoader(dataset1, batch_size=args.batch_size, num_workers=2)
    val_dl   = DataLoader(dataset2, batch_size=args.batch_size, num_workers=2)

    device = "cuda" if torch.cuda.is_available() else "cpu"
    model = CRNNWithAttn().to(device)

    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer, max_lr=1e-3,
        steps_per_epoch=len(train_dl),
        epochs=args.num_epochs,
        anneal_strategy='linear'
    )

    best_val_acc, epochs_no_improve = 0.0, 0
    for epoch in range(1, args.num_epochs + 1):
        train_one_epoch(model, train_dl, optimizer, scheduler, criterion, device, epoch, args.num_epochs)
        val_loss, val_acc = validate(model, val_dl, criterion, device, epoch)

        if val_acc > best_val_acc:
            best_val_acc, epochs_no_improve = val_acc, 0
            ts = TensorSerializer()
            toolbox.save_model(ts.serialize(model.state_dict()), f"{training_dir}/models", metadata={})
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= args.patience:
                print(f"Early stopping at epoch {epoch}")
                break

    print(f"Best Validation Accuracy: {best_val_acc:.4f}")


run(
    main,
    backend=Tractorch(),
    yt_path=training_dir,
    mesh=Mesh(node_count=1, process_per_node=1, gpu_per_process=1, pool_trees=["hyperplane-gpu-h100"]),
    resources=Resources(
        cpu_limit=10,
        memory_limit=32212254720,
    ),
    proxy_stderr_mode=StderrMode.primary,
)
```
